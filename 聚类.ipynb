```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import mglearn

%matplotlib inline
```

```python
data = pd.read_csv('data/seeds.csv')
data.head(5)
```

```python
S = list(data.pop('grain_variety'))
S
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(data)
X
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3).fit(X)
```

```python
data['label'] = km.labels_
data
```

```python
data_x = data[['area','perimeter','compactness','length','width','asymmetry_coefficient','groove_length']]
label = data[['label']]
```

```python
pd.scatter_matrix(data_x, c=km.labels_, figsize=(20,20), cmap=mglearn.cm3)
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.5, min_samples=2).fit(X)
```

```python
pd.scatter_matrix(data_x, c=db.labels_, figsize=(20,20), cmap=mglearn.cm3)
```

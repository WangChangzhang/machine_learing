```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig = plt.figure(num='123', figsize=(10,10))
ax1 = plt.subplot(221)
ax1.plot([1,2,3,4],'r  p')
fig.add_subplot(ax1)
ax2 = plt.subplot(222)
ax2.plot([1,2,3,4],'r  p')
fig.add_subplot(ax2)
ax3 = plt.subplot(223)
ax3.plot([1,2,3,4],'r  p')
fig.add_subplot(ax3)
ax4 = plt.subplot(224)
ax4.plot([1,2,3,4],'r  p')
fig.add_subplot(ax4)
plt.show()

plt.close()
```

```python
import numpy as np
fig = plt.figure(num='scatter', figsize=(10,10), dpi=80)

ax = fig.add_subplot(111)
s1 = ax.scatter(np.random.random(10),np.random.random(10),c='m', marker='x')
s2 = ax.scatter(np.random.random(10),np.random.random(10),c='r', marker='d')
s3 = ax.scatter(np.random.random(10),np.random.random(10),c='g', marker='+')
s4 = ax.scatter(np.random.random(10),np.random.random(10),c='b', marker='*')

ax.legend((s1,s2,s3,s4),('s1','s2','s3','s4',), fontsize=20)
```

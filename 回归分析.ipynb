```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
n_sample = 20

x = np.linspace(0,10 ,20)
x
```

```python
# add ones to array
X = sm.add_constant(x)
X
```

```python
beta = np.array([2,5])
e = np.random.normal(size = n_sample)
```

```python
y = np.dot(X, beta) + e
```

```python
plt.plot(x, y)
```

```python
modle = sm.OLS(y,X)
res = modle.fit()
res.params
```

```python
res.summary()
```

```python
%matplotlib inline
y_ = res.fittedvalues

fig = plt.figure(figsize=(10,10))

plt.plot(x,y,'o',label='original')
plt.plot(x,y_,'r--',label='test')
plt.legend()
```

```python
# y = a + bx + cx^2 a = 2,b=3,c=5

x = np.linspace(0,10, n_sample)
X = np.column_stack((x, x**2))
X = sm.add_constant(X)
X
```

```python
beta2 = np.array([2, 3, 5])
y = np.dot(X, beta2) + e
```

```python
%matplotlib inline
model2 = sm.OLS(y,X)
res2 = model2.fit()
res2.summary()
y2_ = res2.fittedvalues

plt.figure(figsize=(20,20))
plt.plot(x, y, '*', label='data')
plt.plot(x, y2_,'r--', label='test')
plt.legend()
```

```python
# 分类分析 [z1, z2, z3] 取值为 0或1 
# y = 5 + 2x + 2z1 + 3z2 + 6z3
n = 50
beta = np.array([5, 2, 2, 3, 6])
x = np.linspace(0,10, n)
g = np.zeros(n)
g[20:40] = 1
g[40:] = 2
dummy = sm.categorical(g,drop=True)
dummy
```

```python
X = np.column_stack((x,dummy))
X = sm.add_constant(X)
X
```

```python
e = np.random.normal(n)
y = np.dot(X,beta) +e

model = sm.OLS(y,X)
res = model.fit()
```

```python
res.summary()
y_ = res.fittedvalues
```

```python
fig = plt.figure(figsize=(20,20))

plt.plot(x,y,'g*',label='data')
plt.plot(x,y_,'r--',label='test')
```
